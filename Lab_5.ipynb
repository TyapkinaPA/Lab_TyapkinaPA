```python
def gcd(first, second):
    if first > second:
        a = second
    else: 
        a = first
    for i in range(1,a+1):
        if (first % i == 0) and (second % i == 0):
          gcd = i
    return gcd

def lcm(first, second):
    if first > second:
      n = first
    else:
      n = second
    while(True):
      if (n % first == 0) and (n % second == 0):
        return n
      n += 1

def gcd_nums(nums):
    for h in range(1, len(nums)-1):
      nums[h] = gcd(nums[h - 1], nums[h])
    return nums[h]

def lcm_nums(nums1):
    for h in range(1,len(nums) - 1):
      nums1[h] = lcm(nums1[h - 1], nums1[h])
    return nums1[h]

nums = input("Числа: ").split()
nums = list(map(int,nums))
nums1 = nums.copy()
nod = gcd_nums(nums)
nok = lcm_nums(nums1)
print("НОД: ", nod)
print("НОК: ", nok)

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

### Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:
### отрицательных значений (отсортирован по убыванию);

###неотрицательных значений (отсортирован по возрастанию).


```python
def split_numbers(*args):
  p = []
  n = []
  for x in args:
    if x < 0:
      n.append(x)
      n_sort = sorted(n)
      n_sort.reverse()
    else:
      p.append(x)
      p_sort = sorted(p)
  return n_sort, p_sort

print(split_numbers(1, 4, -5, 0, -33))

# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

### Составьте две функции для возведения числа в степень: один из вариантов реализуйте в рекурсивном стиле.

```python
def pow1(value, power):
  if power == 0:
    return 0
  return value ** power

def pow2(value, power):
  if (power == 1):
    return (value)
  if (power != 1):
    return (value * pow2(value, power - 1))

print(pow1(5, 3))
print(pow2(5, 3))

# --------------
# Пример вывода:
#
# 125
# 125
```

## 5.2.17.
### Дано натуральное число. Напишите рекурсивные функции для определения:

### суммы цифр числа;

### количества цифр в числе.

```python
def digits_sum(value):
  if value == 0:
    return 0
  else:
    return (value % 10) + digits_sum(value // 10)

def digits_count(value):
  if value == 1:
    return 1
  else:
    return value ** 0 + digits_count(value // 10)

print(digits_sum(12345))
print(digits_count(12345))

# --------------
# Пример вывода:
#
# 15
# 5
```
